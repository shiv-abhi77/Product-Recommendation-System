```python
df = pd.read_csv("UserData.csv")
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv = CountVectorizer(max_features=5000,stop_words='english')

def getSimiliarity(column):
    vector = cv.fit_transform(column).toarray()
    return cosine_similarity(vector)
```

```python
from sklearn.preprocessing import normalize

def getDistance(column):
    col = normalize([column]).flatten()
    n = col.size
    res = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            res[i][j] = (col[i] - col[j]) ** 2
    return res
```

```python
prod_sim = getSimiliarity(df.prod_name)
cat_sim = getSimiliarity(df.category)
age_dist = getDistance(df.Age)
price_dist = getDistance(df.avg_price)
```

```python
#Calculates final similarity
W_PROD = 7
W_CAT = 10
W_AGE = 2
W_PRICE = 1
similarity = (W_PROD * prod_sim + W_CAT*cat_sim - W_AGE*age_dist - W_PRICE*price_dist) / (W_PROD + W_CAT + W_AGE + W_PRICE)
```

```python
def getSimilar(index):
    print("Users similar to " + df.iloc[index].Name)
    similar = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for ind, score in similar[:10]:
        print(df.iloc[ind])
        print()
```
