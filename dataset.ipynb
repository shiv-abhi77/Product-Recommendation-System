```python
def foo(st):
    key_value_strings = st.split('}, {"')

    all_key_values = []

    for kv_string in key_value_strings:
        key_value = kv_string.split(', ')
        
        key = key_value[0].replace('"key"=>"', '').replace('"', '')
        
        if len(key_value) > 1:
            value = key_value[1].replace('"value"=>"', '').replace('"', '')
            all_key_values.append((key, value))
        else:
            all_key_values.append(("value", key))

    return all_key_values





```

```python
foo(str(new_df['product_specifications'][0]))
```

```python
new_df['prod'] = new_df['product_specifications'].map(lambda x : foo(x))
new_df['prod'][0]
```

```python
def foo2(st):
    all_key_values = []

    for data_tuple in st:
        data_tuple[0].replace('{product_specification=>[{', '').replace('key=>', '').replace('"', '').replace(']}','').strip()
        value = data_tuple[1].replace('"', '').replace('}]}','').replace('{product_specification=>[{', '').replace('value=>', '').strip()
        all_key_values.append((value))

    return all_key_values
```

```python
new_df['prod2'] = new_df['prod'].map(lambda x : foo2(x))
new_df['prod2']=new_df['prod2'].apply(lambda x:[i.replace(" ", "") for i in x])
new_df['prod2'][0]
```

```python
new_df['prod']=new_df['prod2'].apply(lambda x:" ".join(x))
new_df['prod'][0]
```

```python
def removeLooseNumbers(x):
    res = ''
    buf = ''
    for word in x.split(' '):
        buf += word
        is_number = True
        try:
            float(word)
        except ValueError:
            is_number = False
        if not is_number:
            res += ' ' + buf
            buf = ''
    return res

new_df['product_name'] = new_df['product_name'].apply(removeLooseNumbers)
```

```python
new_df['desc'] = new_df['product_name'] + ' ' +  new_df['brand'] + ' ' + new_df['prod']
```

```python
new_df['desc'] = new_df['desc'].apply(lambda x: x.lower())
```

```python
del new_df['brand']
del new_df['prod']
```

```python
def removeStuff(x):
    res = ''
    for c in x:
        if c not in '@"\',/':
            res += c
    return res

new_df['desc'] = new_df['desc'].apply(lambda x: removeStuff(x))
```

```python
new_df.product_category_tree = new_df.product_category_tree.apply(lambda x: x[2:-2].replace('>>', ''))
```

```python
def removeNumbers(x):
    res = ''
    for word in x.split(' '):
        is_number = True
        try:
            float(word)
        except ValueError:
            is_number = False
        if (not is_number) and len(word) < 20:
            res += ' ' + word
    return res
new_df['desc'] = new_df['desc'].apply(removeNumbers)
new_df
```
